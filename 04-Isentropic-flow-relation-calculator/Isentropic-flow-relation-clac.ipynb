```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
```

```python
# Isentropic flow relation calculator
#Constants
gamma = 1.4   #Specific heat ratio for air
M = np.linspace(0.01, 5, 1000)  #Mach number range
```

```python
def temp_ratio(M, gamma):
    T0_T = (1 + (gamma - 1) / 2 * M**2)**(-1)
    return T0_T

def pressure_ratio(M, gamma):
    P0_P = (1 + (gamma - 1) / 2 * M**2)**(-gamma / (gamma - 1))
    return P0_P

def density_ratio(M, gamma):
    rho0_rho = (1 + (gamma - 1) / 2 * M**2)**(-1 / (gamma - 1))
    return rho0_rho

def area_ratio(M, gamma):
    A_Astar = (1 / M) * ((2 / (gamma + 1)) * (1 + (gamma - 1) / 2 * M**2))**((gamma + 1) / (2 * (gamma - 1)))
    return A_Astar
```

```python
#Plotting the results
plt.figure(figsize=(12, 8))
plt.subplot(2, 2, 1)
plt.plot(M, temp_ratio(M, gamma))
plt.title('Total Temperature Ratio (T0/T)')
plt.xlabel('Mach Number (M)')
plt.ylabel('T0/T')
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='M=1 (Sonic)')
plt.grid()
plt.subplot(2, 2, 2)
plt.plot(M, pressure_ratio(M, gamma))
plt.title('Total Pressure Ratio (P0/P)')
plt.xlabel('Mach Number (M)')
plt.ylabel('P0/P')
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='M=1 (Sonic)')
plt.grid()
plt.subplot(2, 2, 3)
plt.plot(M, density_ratio(M, gamma))
plt.title('Total Density Ratio (rho0/rho)')
plt.xlabel('Mach Number (M)')
plt.ylabel('rho0/rho')
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='M=1 (Sonic)')
plt.grid()
plt.subplot(2, 2, 4)
plt.plot(M, area_ratio(M, gamma))
plt.title('Area Ratio (A/A*)')
plt.xlabel('Mach Number (M)')
plt.ylabel('A/A*')
plt.axvline(x=1, color='red', linestyle='--', alpha=0.7, label='M=1 (Sonic)')
plt.grid()
plt.tight_layout()
plt.show()
```

```python
#Ratios
def isentropic_ratios(M, gamma):
    T0_T = temp_ratio(M, gamma)
    P0_P = pressure_ratio(M, gamma)
    rho0_rho = density_ratio(M, gamma)
    A_Astar = area_ratio(M, gamma)
    return T0_T, P0_P, rho0_rho, A_Astar

print("Mach Number (M) | T0/T | P0/P | rho0/rho | A/A*")
for m in [0.01, 0.5, 1, 2, 5]:
    T0_T, P0_P, rho0_rho, A_Astar = isentropic_ratios(m, gamma)
    print(f"{m:14.2f} | {T0_T:.4f} | {P0_P:.4f} | {rho0_rho:.4f} | {A_Astar:.4f}")
```
